```python
male = [d for d in dataset.iterrows() if d[1].loc["sex"] == "m" ]
female = [d for d in dataset.iterrows() if d[1].loc["sex"] == "f" ]
```

```python
def feature(datum):
    a = [0.0] * 40

    # body_type = ["other","average", "fit", "athletic", "thin", "curvy", "a little extra"]
    religion = ["other", "agnosticism", "atheism",  "christianity", "catholicism", "buddhism", "judaism", "hinduism"]
    diet = ["other","anything","vegetarian"]
    drink = ["other", "not at all", "rarely", "socially", "often", "very often"]
    drug = ["never","sometimes","often"]
    smokes = ["other","no","yes","when drinking","sometimes","trying to quit"]
    education = ["other", "dropped out of high school", "working on space camp", "graduated from space camp", "dropped out of space camp", "working on college/university", "dropped out of college/university", "two-year college", "graduated from college/university", "working on masters program", "graduated from masters program", "working on ph.d program", "graduated from ph.d program"]    
    
    for i,b in enumerate(religion): 
        try:
            if b in datum.loc["religion"]: 
                a[i] = 1.0
                break
        except:
            break

    for i,d in enumerate(diet): 
        try:
            if d in datum.loc["diet"]: 
                offset = i + len(religion)
                a[offset] = 1.0
                break
        except:
            break
    
    for i,d in enumerate(drink): 
        if datum.loc["drinks"] == d: 
            offset = i + len(religion) + len(diet) 
            a[offset] = 1.0
            break
    
    for i,d in enumerate(drug): 
        if datum.loc["drugs"] == d: 
            offset = i + len(religion) + len(drink) + len(drink)
            a[offset] = 1.0
            break
        
    for i,d in enumerate(smokes): 
        if datum.loc["smokes"] == d: 
            offset = i + len(religion) + len(drink) + len(drink) + len(drug)
            a[offset] = 1.0
            break

    for i,e in enumerate(education): 
        try:
            if datum.loc["education"] == e or e in datum.loc["education"]: 
                offset = i + len(religion) + len(drink) + len(drink) + len(drug) + len(smokes)
                a[offset] = 1.0
                break
        except:
            break
    

    return [1.0]+a
    
```

```python
def cosine_similarity(X,Y):
    return dot(X,Y) / (norm(X)*norm(Y))
```

```python
def mostSimilar(u, N):
    similarities = []
    user = feature(u)
    # users = usersPerItem[i]
    if u.loc["orientation"] == "straight" and u.loc["sex"] == "m":
        for u2 in female:
            if u2[1].loc["orientation"] != "straight": continue
            user2 = feature(u2[1])
            sim = cosine_similarity(user, user2)
            similarities.append((sim,u2[1]))
    
    if u.loc["orientation"] == "gay" and u.loc["sex"] == "f":
        for u2 in female:
            if u2[1].loc["orientation"] not in ["gay","bisexual"]: continue
            user2 = feature(u2[1])
            sim = cosine_similarity(user, user2)
            similarities.append((sim,u2[1]))
            
    if u.loc["orientation"] == "straight" and u.loc["sex"] == "f":
        for u2 in male:
            if u2[1].loc["orientation"] != "straight": continue
            user2 = feature(u2[1])
            sim = cosine_similarity(user, user2)
            similarities.append((sim,u2[1]))
            
    if u.loc["orientation"] == "gay" and u.loc["sex"] == "m":
        for u2 in male:
            if u2[1].loc["orientation"] not in ["gay","bisexual"]: continue
            user2 = feature(u2[1])
            sim = cosine_similarity(user, user2)
            similarities.append((sim,u2[1]))

    if u.loc["orientation"] == "bisexual":
        opposite = "f" if u.loc["sex"] == "m" else "m"
        for u2 in dataset:
            if u2[1].loc["sex"] == opposite and u2[1].loc["orientation"] == "gay"  or u2[1].loc["sex"] == u.loc["sex"] and u2[1].loc["straight"]: continue
            if u2 == u: continue
            user2 = feature(u2[1])
            sim = cosine_similarity(user, user2)
            similarities.append((sim,u2[1]))
    
    similarities.sort(key=lambda x: x[0], reverse=True)
    return similarities[:N]
```
